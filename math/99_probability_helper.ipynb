### $P(A\mid B) \ne P(B\mid A)$ 　$\mid$ : Given (已知)

Let:

- $A =$ it is a dog
- $B =$ it has four legs

$P(A \mid B) = P(\text{Dog} \mid \text{Has four legs})$ 已知有4隻腳 佢係狗?

This means: given that an animal has four legs, what is the probability that it is a dog?

This probability is much lower, because it could also be a cat, horse, cow, etc.

$P(B \mid A) = P(\text{Has four legs} \mid \text{Dog})$ 已知係狗 有4隻腳?

This means: given that an animal is a dog, what is the probability that it has four legs?

This probability is very high, maybe around $99\%$, because most dogs have four legs, although some may have a disability.

## Bayes' Theorem: Disease Test Example

Bayes' theorem helps us reverse conditional probability:

$$
P(Disease \mid Positive) \ne P(Positive \mid Disease)
$$

Let:

- $D =$ has disease
- $\neg D =$ no disease
- $+ =$ positive test result

---

## Formula

$$
P(D \mid +)
=
\frac{P(+ \mid D)P(D)}{P(+)}
$$

where:

$$
P(+)
=
P(+ \mid D)P(D)
+
P(+ \mid \neg D)P(\neg D)
$$

---

## Example

Suppose:

$$
P(D) = 1\%
$$

$$
P(+ \mid D) = 99\%
$$

$$
P(+ \mid \neg D) = 5\%
$$

Then:

$$
P(\neg D) = 99\%
$$

$$
P(+)
=
(0.99)(0.01) + (0.05)(0.99)
$$

$$
P(+)
=
0.0099 + 0.0495
=
0.0594
$$

Now use Bayes' theorem:

$$
P(D \mid +)
=
\frac{(0.99)(0.01)}{0.0594}
$$

$$
P(D \mid +)
\approx 0.1667
=
16.7\%
$$

---

## Intuition

Even if the test is very accurate:

$$
P(+ \mid D) = 99\%
$$

testing positive does not mean you are 99% likely to have the disease.

Why?

Because the disease is rare.

If we test 10,000 people:

- 100 people have the disease
- 99 of them test positive
- 9,900 people do not have the disease
- 495 of them still test positive

So total positive results:

$$
99 + 495 = 594
$$

Only 99 out of 594 positive results are real disease cases:

$$
\frac{99}{594}
\approx 16.7\%
$$

---

## Key Point

$$
P(+ \mid D)
$$

means:

> If someone has the disease, how likely is the test positive?

$$
P(D \mid +)
$$

means:

> If the test is positive, how likely does the person have the disease?

These are different.

Bayes' theorem helps convert:

$$
P(+ \mid D)
\rightarrow
P(D \mid +)
$$
